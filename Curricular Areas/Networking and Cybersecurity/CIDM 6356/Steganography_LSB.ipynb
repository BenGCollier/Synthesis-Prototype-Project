```python
original_img = "Beauclair_Palace.png"
img_w_message = "Palace_Modified.png"
secret_message = "The idyllic city of Beauclair and its palace at night."
```

```python
from stegano import lsb
secret = lsb.hide(original_img, secret_message)
```

```python
secret.save(img_w_message)
```

```python
hidden_message = lsb.reveal(img_w_message)
```

```python
print(hidden_message)
```
