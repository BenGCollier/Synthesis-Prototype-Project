##Task C. Shipping Logic
Scenario: You are developing code to calculate shipping costs based on a shipping destination. The user needs code that will print the message 'Shipping to UT costs $13.75'  substituting the correct cost for a particular state. Write code to satisfy the following user requirements.

1. A ship_to_state variable has already been coded for you and will have a different value each time you run your code. Do NOT ask the user to input the shipping state.
2. Create a variable called shipping_cost and assigning it a default value of 0.
2. Write an <a href="https://campus.datacamp.com/courses/intermediate-python/logic-control-flow-and-filtering?ex=9">if, elif, else clause</a> to handle the following logic: <br>
If the ship_to_state is TX, NM, CO, or LA, the shipping cost should be 0. <br>
If the ship_to_state is WA, the shipping cost should be 9<br>
If the ship_to_state is AK, the shipping cost should be 16.15<br>
If the ship_to_state is HI, the shipping cost should be 22<br>
All other states should default to 13.75<br>
2. Print the output 'Shipping to TX costs $ 0.00'  substituting the appropriate state abbreviation and cost.  
3. Format the output to include a dollar sign and format the number with two decimal places. Use string interpolation as examples above.

See Figure 3. Sample Output for Task C.<br>
If the ship to state is OK, your code will output the following.
```
'Shipping to OK costs $ 0.00'
```
If the ship to state is WA, your code will output the following.
```
'Shipping to WA costs $ 9.00'
```
If the ship to state is HI, your code will output the following.
```
'Shipping to HI costs $ 22.00'
```
Remember that Dr. Humpherys' prewritten code will select the state randomly from the list of possible states. Run your code multiple times and see that the resulting output changes accordingly.  




````python
import random
possible_states = ['TX', 'NM', 'OK', 'CO', 'LA', 'WA', 'AK', 'HI', 'UT', 'SD', 'KY', 'FL' ]
ship_to_state = random.choice(possible_states) # randomly chooses a state to test many possible ship to locations

shipping_cost = 0

if ship_to_state in ['TX', 'NM', 'CO', 'LA']:
  shipping_cost = 0
elif ship_to_state == 'WA':
  shipping_cost = 9
elif ship_to_state == 'AK':
  shipping_cost = 16.15
elif ship_to_state == 'HI':
  shipping_cost = 22
else:
  shipping_cost = 13.75

print(f'Shipping to {ship_to_state} costs $ {shipping_cost:.2f}')
````
